## 2.1.3: Objects

Создайте класс `Pet` с параметрами `name, pronoun, animal,` и `pet_response`.

Добавьте метод `pet()`, который печатает ответ, состоящий из элементов, например:

` Лейла — кошка. Если ее погладить, Лейла мурчит».`
` Скаут — собака. Если его погладить, Скаут виляет хвостом.`

````python
class Pet:
    # создание экземпляра
    def __init__(a, name, pronoun, animal, pet_response):
        # переменная экземпляра
        a.name = name
        a.pronoun = pronoun
        a.animal = animal
        a.pet_response = pet_response

    # метод, печатающий ответ
    def print_Pet(a):
        print(a.name, " - ", a.animal, ", If you stroke" , a.pronoun, ", ", a.name, a.pet_response, ".", sep="")
````

Вот несколько примеров вызовов:

````python
layla = Pet("Layla", "her", "cat", "purrs")
scout = Pet("Scout", "her", "dog", "wags her tail")
layla.print_Pet()
scout.print_Pet()
````

Выгрузка команд и количества их побед из Википедии

````python
#импортируем библиотеки
import requests
import pandas as pd

#импортируем в датафрейм
url="https://ru.wikipedia.org/wiki/Российская_премьер-лига#Тренеры"
req=requests.get(url)
data=pd.read_html(req.text)
df=data[5]

# Создаем pandas series
ser = df.set_index('Тренер')['Чемпионат']
print(ser)
````

Сортировка по индексу в алфавитном порядке

````python
ser.sort_index(ascending=True)
````
