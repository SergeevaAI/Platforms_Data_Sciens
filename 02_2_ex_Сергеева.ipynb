## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

````python
import pandas as pd
#загрузка файла с компьютера
from google.colab import files
uploaded = files.upload()
````

````python
#словарь для сопоставления буквенных оценок и их числовых значений
gpas = {"A":4, "B":3, "C":2, "D":1}
````

````python
# импортировать библиотеку csv
import csv

# инициализировать массив верхнего уровня
data_values = []

# откройте файл и добавьте строки в виде массивов в data_values
with open('grades.csv') as csvfile:
    filereader = csv.reader(csvfile, delimiter=';', quotechar='"')
    for row in filereader:
        # Добавление среднего балла
        row.append(gpas[row[1].strip()])
        data_values.append(row)
````

````python
# Создание нового DataFrame
df = pd.DataFrame(data_values, columns=['Имя', 'Буквенная оценка', 'Числовая оценка'])

# Запись данных в новый файл grades_gpa.csv
df.to_csv('grades_gpa.csv', index=False)
````

 * Создайте новый фрейм данных на основе набора данных hit_albums, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

````python
#считаем количество альбомов для каждого артиста
hit_albums2=hit_albums
mask = hit_albums2["Artist"].value_counts().loc[lambda x: x >= 2].reset_index()

#переименовываем столбцы
mask.rename(columns={"index": "Artist", "Artist": "Count album"}, inplace = True)
mask
````

````python
#объединение таблиц
hit_albums3=pd.merge(hit_albums2, mask, on='Artist')

#удаление лишнего столбца
hit_albums3.drop('Count album_x', axis=1, inplace=True)

#переименование столбца
hit_albums3.rename(columns={"Count album_y": "Count album"}, inplace = True )
hit_albums3
````

В результате вывелась таблица с артистами, у которых как минимум 2 альбома, и количеством альбомов для каждого артиста.

Больше всего альбомов у Michael Jackson (5).

* Создайте новый фрейм данных, содержащий общую сумму всех сертифицированных продаж за каждый год.

````python
#группировка по году
hit_albums4 = hit_albums['Certified sales (millions)'].groupby(hit_albums['Released']).sum()

#сбрасывание индексации
hit_albums4.reset_index()
````

Больше всего продаж (144.5) в 1987 году, что обусловлено выходом альбомов сразу нескольких известных личностей, одна из которых самая продаваемая по проведенному выше анализу (Michael Jackson	), к тому же в тот период Европа переживала расцвет дэнс-поп музыки.
